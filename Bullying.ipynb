```
bullying = pd.read_csv('bullying.csv')

bullying = bullying.drop('Unnamed: 0', axis='columns')

bullying
```

```
labels = [2009, 2011, 2013, 2015]
r_disability = bullying['Reported Disability']
d_disability = bullying['Disciplined Disability']
r_sex = bullying['Reported Sex']
d_sex = bullying['Disciplined Sex']
r_race = bullying['Reported Race']
d_race = bullying['Disciplined Race']
x = np.arange(len(labels))  # the label locations
width = .35  # the width of the bars
```

```
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')

rects1A = ax.bar(x-(width/2), r_disability, width, 
                 label='Reported- Disability', color = '#4771B6')
rects1B = ax.bar(x+(width/2), d_disability, width, 
                 label='Disciplined- Disability', color = '#2E5C97')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Incidents')
ax.set_title('Reported vs Disciplined Bullying Incidents- Disability')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```

```
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')

rects1A = ax.bar(x-(width/2), r_sex, width, 
                 label='Reported- Sex', color = '#E6A438')
rects1B = ax.bar(x+(width/2), d_sex, width, 
                 label='Disciplined- Sex', color = '#C58317')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Incidents')
ax.set_title('Reported vs Disciplined Bullying Incidents- Sex')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```
