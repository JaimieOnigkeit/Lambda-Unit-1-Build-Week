```
df = pd.read_csv('school_demo.csv')
df = df.drop('Unnamed: 0', axis= 'columns')
df = df.drop([9,10,11], axis= 'rows')

df
```

```
dfT = df.T

dfT = dfT.rename(columns = {0 : 'American Indian or Alaskan Native', 
                            1:'Asian or Pacific Islander', 
                            2:'Asian',
                            3:'Hawaiian/Pacific Islander',
                            4:'Hispanic',
                            5:'Black', 6:'White',
                            7:'Two or More Races',
                            8:'Total'})
dfT = dfT.drop('Category', axis='rows')
dfT = dfT.drop('2000', axis='rows')
dfT = dfT.drop('2004', axis='rows')
dfT = dfT.drop('2006', axis='rows')
dfT = dfT.drop('Asian or Pacific Islander', axis='columns')

dfT
```

```
data1 = dfT['American Indian or Alaskan Native']
data3 = dfT['Asian']
data4 = dfT['Hawaiian/Pacific Islander']
data5 = dfT['Hispanic']
data6 = dfT['White']
data7 = dfT['Two or More Races']
data8 = dfT['Total']
data9 = dfT['Black']
```

```
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')

ax.plot(data8, label='Total', color='#4771A6')
ax.plot(data6, label='White', color='#7F9606')
ax.plot(data5, label='Hispanic', color='#A13EA5')
ax.plot(data3, label='Asian', color='#C58317')
ax.plot(data4, label='Pacific Islander or Native Hawaiian', color='#2E5C97')
ax.plot(data9, label='Black', color='#922596')
ax.plot(data1, label='American Native or Native Alaskan', color='#C3DF36')
ax.plot(data7, label='Two or More Races', color='#C3DF36')

ax.grid(linestyle="--", linewidth=0.5, color='#7C1080', zorder=-10)

ax.legend()

plt.show()
```
