```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Utah Demographics

```
#Read in Utah demographics data
df = pd.read_csv('https://raw.githubusercontent.com/JaimieOnigkeit/Lambda-Unit-1-Build-Week/master/utah_demograhics2.csv')

df
```

## Demographics visual

```
#Transpose the data and rename the columns
dfT = df.T

dfT = dfT.rename(columns = {0 : 'Total', 
                            1:'Non Hisp Total', 
                            2:'NH White',
                            3:'Minority',
                            4:'Hispanic',
                            5:'Black',
                            6:'Asian', 
                            7:'American Indian or Alaskan Native',
                            8:'Pacific Islander or Native Hawaiian',
                            9:'NH Multi'})

dfT = dfT.drop('Unnamed: 0', axis='rows')

dfT
```

```
#Create data variables for each column
data1 = dfT['American Indian or Alaskan Native']
data3 = dfT['Asian']
data4 = dfT['Pacific Islander or Native Hawaiian']
data5 = dfT['Hispanic']
data6 = dfT['NH White']
data7 = dfT['NH Multi']
data8 = dfT['Total']
data9 = dfT['Black']
```

```
#Create the graph
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')

ax.plot(data8, label='Total', color='#C58317')
ax.plot(data6, label='White', color='#7F9606')
ax.plot(data5, label='Hispanic', color='#A13EA5')
ax.plot(data3, label='Asian', color='#4771A6')
ax.plot(data4, label='Pacific Islander or Native Hawaiian', color='#2E5C97')
ax.plot(data9, label='Black', color='#922596')
ax.plot(data1, label='American Native or Native Alaskan', color='#C3DF36')
ax.plot(data7, label='Two or More Races', color='#E6A438')

ax.grid(linestyle="--", linewidth=0.5, color='#7C1080', zorder=-10)
ax.set_title('Demographic Changes in Utah Population')

ax.legend()

plt.show()
```

```
#Create graph focusing on minority growth
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')

ax.plot(data5, label='Hispanic', color='#A13EA5')
ax.plot(data3, label='Asian', color='#4771A6')
ax.plot(data4, label='Pacific Islander or Native Hawaiian', color='#2E5C97')
ax.plot(data9, label='Black', color='#922596')
ax.plot(data1, label='American Native or Native Alaskan', color='#C3DF36')
ax.plot(data7, label='Two or More Races', color='#E6A438')

ax.grid(linestyle="--", linewidth=0.5, color='#7C1080', zorder=-10)
ax.set_title('Minority Demographic Changes in Utah Population')

ax.legend()

plt.show()
```

## School Demographics


```
#Read in school enrollment data
enrollment = pd.read_csv('https://raw.githubusercontent.com/JaimieOnigkeit/Lambda-Unit-1-Build-Week/master/Enrollment_Data2.csv')

enrollment
```

```
#Eliminate unnecessary columns
enrollment = enrollment.drop('Lea State', axis = 'columns')
enrollment = enrollment.drop('School State', axis = 'columns')
enrollment = enrollment.drop('ID', axis = 'columns')
enrollment = enrollment.drop('Category', axis = 'columns')
enrollment = enrollment.drop('SWD (IDEA-Eligible)', axis='columns')
enrollment = enrollment.drop('SWD (Section 504 only)', axis='columns')
enrollment = enrollment.drop('LEP', axis = 'columns')

enrollment.head()
```

```
#Check for null values
enrollment.isnull().sum()
```

```
#Replace null values
enrollment.fillna(0, inplace= True)
```

## Bullying Visualization

```
#Create graph
labels = [2009, 2011, 2013, 2015]
r_race = EE_final['Reported Race']
d_race = EE_final['Disciplined Race']
x = np.arange(len(labels))  # the label locations
width = .35  # the width of the bars

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
fig.set_facecolor(color='White')


rects1A = ax.bar(x-(width/2), r_race, width, 
                 label='Reported- Race', color = '#C3DF36')
rects1B = ax.bar(x+(width/2), d_race, width, 
                 label='Disciplined- Race', color = '#7F9606')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Incidents')
ax.set_title('Reported vs Disciplined Bullying Incidents- Race')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```
