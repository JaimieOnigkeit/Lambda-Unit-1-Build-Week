```
colors = ['#4771A6', '#7F9606', '#7C1080', '#C58317', '#2E5C97', '#922596', '#C3DF36', '#E6A438', '#A13EA5']
```

```
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['White', 'Hispanic', 'Black', 'American Native or Alaskan Native', 'Asian', 'Pacific Islander or Native Hawaiian', 'Two or More']
sizes = [(utah_demo['Census 2010'][1]), (utah_demo['Census 2010'][4]), 
         (utah_demo['Census 2010'][5]),(utah_demo['Census 2010'][6]),
         (utah_demo['Census 2010'][7]), (utah_demo['Census 2010'][8]),
         (utah_demo['Census 2010'][9])]
explode = (0, 0.1, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.set_title('2010', fontsize=22)
fig1.set_facecolor(color='White')

fig1.set_figheight(10)
fig1.set_figwidth(10)
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=10, textprops={'fontsize':14, 'fontweight':'bold'})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
